```python
def czy_pierwsza(potencjalna_pierwsza):
    for y in range(2, potencjalna_pierwsza):
        if potencjalna_pierwsza % y == 0:
            return False
    return True
```

```python
liczba = 7

for x in range(2, liczba):
    if czy_pierwsza(x):
        print(x)
```
