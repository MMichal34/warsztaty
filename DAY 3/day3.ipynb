```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
# Definicja funkcji
def complicated_function(x):
    return np.sin(x) * np.exp(-0.2 * x)

# Tworzenie danych do wykresu
x = np.linspace(0, 10, 100)  # Zakres wartości x
y = complicated_function(x)  # Obliczanie wartości funkcji dla każdego x

# Tworzenie wykresu
plt.plot(x, y, linewidth=2)

# Dodawanie etykiet i tytułu
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wizualizacja skomplikowanej funkcji')

# Wyświetlanie wykresu
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definicja funkcji dwuwymiarowej
def complicated_function(x, y):
    return np.sin(x) * np.cos(y)

# Tworzenie siatki punktów w 2D
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)  # Tworzenie siatki X i Y

# Obliczanie wartości funkcji dla każdego punktu siatki
Z = complicated_function(X, Y)

# Tworzenie wykresu 2D
plt.imshow(Z, cmap='hot', extent=[-np.pi, np.pi, -np.pi, np.pi])
plt.colorbar()

# Dodawanie etykiet i tytułu
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wizualizacja skomplikowanej funkcji 2D')

# Wyświetlanie wykresu
plt.show()
```

```python
import seaborn as sns

def cost(state) -> float:
    x, y = state
    return -5*x - 8*y

matrix = [[cost((x, y)) for x in range(-10, 11)] for y in range(-10, 11)]
sns.heatmap(matrix)
```
